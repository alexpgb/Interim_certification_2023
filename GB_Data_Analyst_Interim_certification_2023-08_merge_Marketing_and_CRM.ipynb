```python
if 'tag' in df_data_marketing.columns.to_list():
    df_data_marketing.drop(['tag'], axis= 1 , inplace= True )    
df_data_marketing.insert (loc= len(df_data_marketing.columns), column='tag',
                            value=[0 for i in range(len(df_data_marketing))])
if 'tag' in df_data_crm.columns.to_list():
    df_data_crm.drop(['tag'], axis= 1 , inplace= True )    
df_data_crm.insert (loc= len(df_data_crm.columns), column='tag',
                            value=[0 for i in range(len(df_data_crm))])
```

```python
df_data_marketing_gr = df_data_marketing[['Client_ID','ID']].groupby('Client_ID').\
count().reset_index().rename(columns={'ID':'count'})
```

```python
df_merge = pd.DataFrame(columns=['ID_M','ID_C'])
for i in range(len(df_data_marketing_gr)):    
    exp_f_marketing = df_data_marketing['Client_ID'] == df_data_marketing_gr.iloc[i]['Client_ID']#матрица фильтрации marketing 
    exp_f_crm = df_data_crm['Client_ID'] == df_data_marketing_gr.iloc[i]['Client_ID']#матрица фильтрации crm
    if df_data_marketing_gr.iloc[i]['count'] == 1:
        df_merge.loc[len(df_merge.index)] = [df_data_marketing[exp_f_marketing]['ID'].values[0],
                                             df_data_crm[exp_f_crm]['ID'].values[0]]
        df_data_marketing.loc[exp_f_marketing,['tag']] = 1
        df_data_crm.loc[exp_f_crm,['tag']] = 1
    else:
        # df_data_crm_f = df_data_crm[exp_f_crm]
        ic = 0
        rc = np.NaN     
        for im, rm in df_data_marketing[exp_f_marketing].iterrows():
            ic = 0
            rc = np.NaN
            # print(im, rm['ID'])
            if rm['Conversion'] == 0:
                for ic, rc in df_data_crm[(exp_f_crm) & (df_data_crm['tag'] == 0) & 
                                         (df_data_crm['Tag_ComingToDealer'] == 0)].iterrows():
                    df_merge.loc[len(df_merge.index)] = [rm['ID'], rc['ID']]
                    df_data_marketing.at[im, 'tag'] = 1
                    df_data_crm.at[ic, 'tag'] = 1
                    break
            else:  # нужно искать первую попавшуюся запись с визитом к дилеру 
                for ic, rc in df_data_crm[(exp_f_crm) & (df_data_crm['tag'] == 0)].iterrows():
                    if rc['Tag_ComingToDealer'] == 1: 
                        break # если эта запись 
                df_merge.loc[len(df_merge.index)] = [rm['ID'], rc['ID']]
                df_data_marketing.at[im, 'tag'] = 1
                df_data_crm.at[ic, 'tag'] = 1
```
